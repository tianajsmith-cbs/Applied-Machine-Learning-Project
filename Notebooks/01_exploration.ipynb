```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/Users/rathsusan/Documents/Applied-Machine-Learning-Project/Data/Raw/online_shoppers_intention 2.csv")
df.head()
```

```python
print(df.info())
df.describe(include='all')
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("Column names:", df.columns.tolist())
```

```python
# check missing values
df.isnull().sum()
# unique values in categorical columns for none numerical ones
for col in ['Month', 'VisitorType', 'Weekend', 'Revenue']:
    if col in df.columns:
        print(f"\n{col} unique values:\n", df[col].unique())
```

```python
for col in ['OperatingSystems', 'Browser', 'Region', 'TrafficType']:
    if col in df.columns:
        print(f"\nColumn: {col}")
        print(f"Unique values: {df[col].unique()}")
        print(f"Number of unique categories: {df[col].nunique()}")
        print("\nCategory frequencies:")
        print(df[col].value_counts())
        print("-" * 50)
```

```python
# numeric distributions
df.hist(bins=30, figsize=(14,10))
plt.suptitle("Numeric Distributions", y=1.02)
plt.show()
```

```python
# showing how many times a revenue resuted in either a purchase or not 
sns.countplot(x="Revenue", data=df)
plt.title("Revenue Distribution (Target Variable)")
plt.show()
```

```python
# conversion rate by Month
if 'Month' in df.columns:
    conv = df.groupby('Month')['Revenue'].mean()
    conv.plot(kind='bar', figsize=(8,4))
    plt.title("Conversion Rate by Month")
    plt.ylabel("Mean Revenue (Conversion Rate)")
    plt.show()
```

```python
# conversion rate by VisitorType
if 'VisitorType' in df.columns:
    conv_type = df.groupby('VisitorType')['Revenue'].mean().sort_values(ascending=False)
    conv_type.plot(kind='bar', figsize=(6,4))
    plt.title("Conversion Rate by Visitor Type")
    plt.ylabel("Mean Revenue (Conversion Rate)")
    plt.show()
```
