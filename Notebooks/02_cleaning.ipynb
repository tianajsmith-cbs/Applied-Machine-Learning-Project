```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/Users/rathsusan/Documents/Applied-Machine-Learning-Project/Data/Raw/online_shoppers_intention 2.csv")
df.head()
```

```python
df.isnull().sum
```

```python
df.dtypes
```

```python
# Drop Month feature 
df = df.drop(columns=['Month'])
```

```python
df.head()
```

```python
#drop pagevalues to prevent dataleakage
df = df.drop(columns=['PageValues'])
df.head()
```

```python
categorical_cols = ['OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType']

# Apply one-hot encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.columns
```

```python
df.dtypes
```

```python
output_path = "../Data/Processed/online_shoppers_clean.csv"
df.to_csv(output_path, index=False)
```
